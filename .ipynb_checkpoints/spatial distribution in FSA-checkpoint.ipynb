```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd

# Load your wildfire data
fire_data = pd.read_csv('/Applications/fire.csv')

# Create a GeoDataFrame from wildfire data
geometry = [Point(xy) for xy in zip(fire_data['fire_location_longitude'], fire_data['fire_location_latitude'])]
wildfire_gdf = gpd.GeoDataFrame(fire_data, geometry=geometry, crs='EPSG:4326')

# Load FSA boundaries data
fsa_boundaries = gpd.read_file('/Applications/fire.shp')

# Set CRS for both GeoDataFrames
wildfire_gdf = wildfire_gdf.set_crs('EPSG:4326')
fsa_boundaries = fsa_boundaries.set_crs('EPSG:4326')

# Spatial join to associate wildfires with FSA boundaries
merged_data = gpd.sjoin(wildfire_gdf, fsa_boundaries, how='left', op='within')

# Display the merged GeoDataFrame
print(merged_data)

# Check CRS
print(wildfire_gdf.crs)
print(fsa_boundaries.crs)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load FSA boundaries data
fsa_boundaries = gpd.read_file('/Applications/fire.shp')

# Plot FSA boundaries
fsa_boundaries.plot()

# Show the plot
plt.show()
```
