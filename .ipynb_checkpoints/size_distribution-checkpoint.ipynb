```python
import pandas as pd

# Load your CSV file into a DataFrame
wildfire_data = pd.read_csv('/Applications/fire.csv')

# Display the first few rows of the DataFrame
wildfire_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(wildfire_data['current_size'], bins=50, kde=True, color='skyblue', log_scale=(True, True))
plt.title('Distribution of Wildfire Sizes (Log Scale)')
plt.xlabel('Wildfire Size (Log Scale)')
plt.ylabel('Frequency (Log Scale)')
plt.show()

```
